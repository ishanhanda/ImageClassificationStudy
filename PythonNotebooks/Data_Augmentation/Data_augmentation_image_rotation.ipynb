```python
import cv2 
import matplotlib.pyplot as plt
import numpy
import os
from os import listdir
from os.path import isfile, join

# beaver_training = '/Users/ishanhanda/Documents/NYU_Fall16/Comp_Vision/Project/ProjectWorkspace/DataSets/MyDataSets/Augmentation/Beaver_train_aug/'
negative_training = '/Users/ishanhanda/Documents/NYU_Fall16/Comp_Vision/Project/ProjectWorkspace/DataSets/MyDataSets/Augmentation/Negative_train_aug/'
# rotated_Folder = '/Users/ishanhanda/Documents/NYU_Fall16/Comp_Vision/Project/ProjectWorkspace/DataSets/MyDataSets/Augmentation/Rotated/'
rotated_Folder_neg = '/Users/ishanhanda/Documents/NYU_Fall16/Comp_Vision/Project/ProjectWorkspace/DataSets/MyDataSets/Augmentation/Rotated_neg/'
allImages = [f for f in listdir(negative_training) if isfile(join(negative_training, f))]


angles = numpy.arange(18, 360, 18)

for image in allImages: 
    name = (os.path.splitext(image)[0])
    img = cv2.imread(negative_training + image,0)
    rows,cols = img.shape
    for angle in angles:
        M = cv2.getRotationMatrix2D((cols/2,rows/2),angle,1)
        dst = cv2.warpAffine(img,M,(cols,rows))
        filename = "{}_{}.JPEG".format(name, angle)
        print(filename)
        cv2.imwrite(rotated_Folder_neg + filename,dst)
```
